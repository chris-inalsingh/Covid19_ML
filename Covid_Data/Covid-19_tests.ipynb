```python
#Dependencies
import pandas as pd
import os
```

```python
#path to collect csv
file_path = os.path.join('Covid-19_tests.csv')
#read in csv
C19_df = pd.read_csv(file_path)
C19_df.head(5)
#0=False, 1=True
```

```python
C19_df['test_indication'].unique()
```

```python
#replace 3 word value to 1 word value
C19_df2 = C19_df.replace('Contact with confirmed', 'Contact')
C19_df2['test_indication'].unique()
```

```python
C19_df3 = C19_df2.rename(columns={'corona_result':'covid_result'})
C19_df3
```

```python
C19_df3=C19_df3.dropna()
C19_df3
```

```python
C19_df3=C19_df3[C19_df3.covid_result != "other"]
C19_df3
```

```python
C19_df3['covid_result'].unique()
```

```python
C19_df3.dtypes
```

```python
selected_features = C19_df3[["cough","fever","sore_throat","shortness_of_breath","head_ache","age_60_and_above", "gender","test_indication"]]
```

```python
selected_features.loc[selected_features['test_indication'] == 'Abroad']
```

```python
selected_features["age_60_and_above"]=selected_features.age_60_and_above.astype('category')
selected_features["gender"]=selected_features.gender.astype('category')
selected_features["test_indication"]=selected_features.test_indication.astype('category')
```

```python
cat_columns = selected_features.select_dtypes(['category']).columns
cat_columns
```

```python
selected_features.dtypes
```

```python
selected_features[cat_columns] = selected_features[cat_columns].apply(lambda x: x.cat.codes)
```

```python
selected_features.head()
```

```python
selected_features.loc[selected_features['test_indication'] == 2]
```

```python
selected_features['age_60_and_above'].unique()
#0=No, 1=Yes
```

```python
selected_features['gender'].unique()
#0=Female, 1=Male
```

```python
selected_features['test_indication'].unique()
#0=Abroad, 1=Contact, 2=Other
```

```python
target=C19_df3["covid_result"]
target.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(selected_features, target, random_state=42, stratify=target)
```

```python
X_train.head()
```

```python
# Scale your data
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
#Train model
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model.fit(X_train_scaled, y_train)
```

```python
print(f"Training Data Score: {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")
```

```python
# Calculate classification report for SVC
from sklearn.metrics import classification_report
predictions = model.predict(X_test_scaled)
print(classification_report(y_test, predictions,
                            target_names=["negative", "positive"]))
```

```python
#Save the model
import joblib
filename = 'SVC_model.sav'
joblib.dump(model, filename)
```
